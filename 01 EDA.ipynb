#### 10/11 application_{train|test}.csv

This is the main table, broken into two files for Train (with TARGET) and Test (without TARGET).  

Static data for all applications.  

One row represents one loan in our data sample.  

これがメインテーブルで、Train (TARGET付き)とTest (TARGET無し)の2つのファイルに分かれています。 　　

すべてのアプリケーションの静的データ。

1 つの行は、データサンプルの 1 つのローンを表します。 

#### 20 bureau.csv

All client's previous credits provided by other financial institutions that were reported to Credit Bureau (for clients who have a loan in our sample).  

For every loan in our sample, there are as many rows as number of credits the client had in Credit Bureau before the application date.  

他の金融機関から信用情報機関に報告された全てのお客様の過去の債権（当社サンプルで融資を受けているお客様の場合）。 

私たちのサンプルのすべてのローンについては、クライアントが申請日前にクレジットビューローで持っていたクレジットの数のように多くの行があります。   

#### 30 bureau_balance.csv

Monthly balances of previous credits in Credit Bureau.  

This table has one row for each month of history of every previous credit reported to Credit Bureau  

– i.e the table has (#loans in sample * # of relative previous credits * # of months where we have some history observable for the previous credits) rows.

クレジットビューローの過去のクレジットの毎月の残高。 　　

この表には、信用情報機関に報告された過去のクレジットの履歴の各月ごとに1つの行があります。　　

私たちはいくつかの歴史を持っている月の＃我々は以前のクレジットの行のために観測可能ないくつかの歴史を持っている場所の相対的な以前のクレジットの＃＃＃サンプル*のローンを持っています。  

#### 40 POS_CASH_balance.csv
#### 50 credit_card_balance.csv

Monthly balance snapshots of previous credit cards that the applicant has with Home Credit.  

This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample  

– i.e. the table has (#loans in sample * # of relative previous credit cards * # of months where we have some history observable for the previous credit card) rows.

申込者がホームクレジットで持っている過去のクレジットカードの毎月の残高スナップショット。 　

この表は、サンプルのローンに関連するホームクレジット（消費者金融とキャッシングローン）の過去のクレジットの履歴の各月ごとに1つの行を持っています。　　

すなわち、テーブルは、（サンプル内のローン * # は前のクレジットカードに関係する）行を持っています。　　

#### 60 previous_application.csv

All previous applications for Home Credit loans of clients who have loans in our sample.  

There is one row for each previous application related to loans in our data sample.  

当社のサンプルで融資を受けているお客様の過去のホームクレジットローンのお申込みはすべて対象となります。  

私たちのデータサンプルのローンに関連する各過去のアプリケーションのための1つの行があります。  

#### 70 installments_payments.csv

Repayment history for the previously disbursed credits in Home Credit related to the loans in our sample.  

There is a) one row for every payment that was made plus b) one row each for missed payment.  

One row is equivalent to one payment of one installment OR one installment corresponding to one payment of one previous Home Credit credit related to loans in our sample.  

本サンプルのローンに関連するホームクレジットの過去の返済履歴。  

a) 支払いがあった分に1列＋b) 支払いがなかった分に1列ずつあります。  

1 行は、1 回払いの 1 回払いまたは 1 回払いの 1 回払いに対応する 1 回払いの 1 回払いに相当する。  

#### 80 HomeCredit_columns_description.csv

This file contains descriptions for the columns in the various data files.  

このファイルには、さまざまなデータファイルのカラムの説明が含まれています。
